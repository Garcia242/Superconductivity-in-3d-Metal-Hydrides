```python
#program to plot DOS 

#need to manually replace the fermi energy for each materials for this to work

import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
%matplotlib inline

# load data
data = np.loadtxt("/Users/guilhermegarcia/GLCGarcia documents /University /3rd year 2023 /Summer Development Internship /DOS_anaylis/Fe/FeH_dos.dat")
data2 = np.loadtxt('/Users/guilhermegarcia/GLCGarcia documents /University /3rd year 2023 /Summer Development Internship /DOS_anaylis/Fe/Fe3H8_dos.dat')
data_FeH3 = np.loadtxt("/Users/guilhermegarcia/GLCGarcia documents /University /3rd year 2023 /Summer Development Internship /DOS_anaylis/Fe/FeH3_dos.dat")
print(data[:,0])

energy1 = data[:,0]
dos1 = data[:,1]
idos1 = data[:,2]

energy2 = data2[:,0]
dos2 = data2[:,1]
idos2 = data2[:,2]

energy_FeH3 = data_FeH3[:,0]
dos_FeH3 = data_FeH3[:,1]
idos_FeH3 = data_FeH3[:,2]

# make plot
plt.figure(figsize = (12, 6))
plt.plot(energy1, dos1, linewidth=0.75, color='red')
plt.yticks([])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS')
plt.axvline(x=188.853, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
#plt.xlim(-6, 16)
plt.ylim(0, )
plt.fill_between(energy1, 0, dos1, where=(energy1 < 188.853), facecolor='red', alpha=0.25)
#plt.text(6, 1.7, 'Fermi energy',  rotation=90)
plt.title("FeH")
plt.show()

plt.figure(figsize = (12, 6))
plt.plot(energy2, dos2, linewidth=0.75, color='red')
plt.yticks([])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS')
plt.axvline(x=126.718, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
#plt.xlim(-6, 16)
plt.ylim(0, )
plt.fill_between(energy2, 0, dos2, where=(energy2 <  126.718), facecolor='red', alpha=0.25)
plt.title("Fe3H8")
#plt.text(6, 1.7, 'Fermi energy',  rotation=90)
plt.show()

plt.figure(figsize = (12, 6))
plt.plot(energy_FeH3, dos_FeH3, linewidth=0.75, color='red')
plt.yticks([])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS')
plt.axvline(x=127.116, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
#plt.xlim(-6, 16)
plt.ylim(0, )
plt.fill_between(energy_FeH3, 0, dos_FeH3, where=(energy_FeH3 < 127.116), facecolor='red', alpha=0.25)
plt.title("FeH3")
#plt.text(6, 1.7, 'Fermi energy',  rotation=90)
plt.show()
```
